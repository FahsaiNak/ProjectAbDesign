```python
#take in atom df and a query_value ()
def get_letter_to_extract(head_df, query_value):
    
    head_df.drop_duplicates(subset=['entry'], inplace=True)
    chain_indices = head_df["entry"].str.find(query_value+'CHAIN=')+len(query_value+'CHAIN=')
    first_index = (chain_indices > len(query_value+'CHAIN=')).idxmax()
    if first_index == 0:
        return None
    first_value = chain_indices[first_index]
    letter_to_extract = head_df['entry'].iloc[first_index][first_value:first_value+1]

    return letter_to_extract
```

```python
def getCDRSeq(head_df, atom_df, CDR_type, pdb_id, CDR_file_path):
    if CDR_type== 'H':
        CDR_pos = CDR_Hpos
    elif CDR_type == 'L':
        CDR_pos = CDR_Lpos
    parser = PDBParser(QUIET=True)
    

    #get the letter of the H or L chain in this pdb entry (A, B, C, etc.)
    letter_to_extract = get_letter_to_extract(head_df, CDR_type)
    
    if letter_to_extract is None:
        print(f'No {CDR_type} chain in file')
        return None

    chain_df = atom_df[atom_df.chain_id == letter_to_extract].reset_index(drop=True)

    for ind, val in CDR_pos.items():

        CDR_dict['chain'].append(letter_to_extract)
        CDR_dict['CDR'].append(ind)
        CDR_dict['PDB_id'].append('insertid') #to change
        start_residue = val[0]
        end_residue = val[1]

        start_residue_df = chain_df[chain_df.residue_number==start_residue]
        if len(start_residue_df)==0:
            print(f'Incomplete {ind} in file')
            return None
        start_residue_atom_idx = start_residue_df.index.values[0]
        end_residue_df = chain_df[chain_df.residue_number==end_residue]
        if len(end_residue_df)==0:
            print(f'Incomplete {ind} in file')
            return None
        end_residue_atom_idx = end_residue_df.index.values[-1]

        CDR_df = chain_df.iloc[start_residue_atom_idx:end_residue_atom_idx+1].reset_index(drop=True)

        pdb_to_save = PandasPdb().read_pdb(chothia_file_path+'/'+pdb_id+'.pdb')

        pdb_to_save.df['ATOM'] = pd.merge(pdb_to_save.df['ATOM'], CDR_df, how='inner')

        pdb_to_save.to_pdb(path=CDR_file_path +'/' +pdb_id+'_'+ind+'.pdb',
                                            records=['ATOM', 'OTHERS', 'ANISOU'],
                                            gz=False,
                                            append_newline=True)
    
#     structure = parser.get_structure(pdb_id, CDR_file_path +'/' +pdb_id+'_'+ind+'_'+letter_to_extract+'.pdb') 
#     for model in structure:

#         for chain in model:
#             seq= []
#             for residue in chain:
#                 seq.append(d3to1[residue.resname])
#             CDR_seq=''.join(seq)
#     CDR_dict['sequence'].append(CDR_seq)
#     print('Completed')
```

```python
def searchBreakChain(reindex_df):
    for i,line in enumerate(reindex_df.line_idx):
        if i == 0:
            before_line = line
            continue
        if before_line+1!=line:
            break_ind = i
            break
        before_line=line
        if i== reindex_df.index[-1]:
            break_ind = i+1
    return break_ind
def checkConsecutive(l):
    return sorted(l)==list(range(min(l),max(l)+1))
```
